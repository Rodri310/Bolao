# Visualização e Consolidação dos Dados do Bolão

Este notebook importa as funções do script `Data_get.py` para extrair, processar e consolidar as tabelas de dados do bolão (`dim_jogos`, `dim_membros` e `fato_palpites`) em um único DataFrame desnormalizado, ideal para ser consumido por aplicações do **Plotly Dash**.

```python
from Data_get import obter_dados_consolidados, extrair_dados_bolao
import pandas as pd
```

## 1. Carregando a Tabela Única Consolidada

A função `obter_dados_consolidados` lê o arquivo HTML e realiza os merges (joins) automaticamente usando o Pandas (`fato_palpites` mesclado com `dim_membros` e `dim_jogos`).

```python
caminho_html = 'auditoria-guerreiros (1).html'
df_consolidado = obter_dados_consolidados(caminho_html)
```

## 2. Inspeção da Tabela Única (para uso no Plotly Dash)

Esta tabela possui todos os palpites, com o nome do participante, informações e dados do jogo correspondente na mesma linha.

```python
print(f"Total de palpites/linhas: {len(df_consolidado)}")
print("Colunas disponíveis:", df_consolidado.columns.tolist())
df_consolidado.head()
```
